# 1. load files

```python
def read_csv(filepath: Path) -> pd.DataFrame:
    df = pd.read_csv(
        filepath_or_buffer = filepath,
        sep = ',',
        engine = 'python',
        encoding = 'utf-8',
        index_col = None
    )
    return df
```

```python
def read_ecoinvent_txt(filepath: Path) -> pd.DataFrame:
    df = pd.read_csv(
        filepath_or_buffer = filepath,
        sep = '\t',
        engine = 'python',
        encoding = 'utf-8',
        index_col = None
    )
    return df
```
