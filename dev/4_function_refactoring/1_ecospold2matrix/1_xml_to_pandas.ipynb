```python
def create_activity_df(
    path_file_activity_names: PosixPath,
    path_file_activity_index: PosixPath,
    dict_column_names_activity_index: dict = {
        'id': 'activityId',
        'activityNameId': 'activityNameId',
        'specialActivityType': 'activityType',
        'startDate': 'startDate',
        'endDate': 'endDate'
    },
    dict_column_names_activity_names: dict = {
        'id': 'activityNameId',
        'name': 'name'
    },
) -> pd.DataFrame:
    """_summary_

    Args:
        path_file_activity_names (PosixPath): _description_
        path_file_activity_index (PosixPath): _description_
        dict_column_names_activity_names (_type_, optional): _description_. Defaults to { 'id': 'activityId', 'activityNameId': 'activityNameId', 'specialActivityType': 'activityType', 'startDate': 'startDate', 'endDate': 'endDate' }.

    Returns:
        pd.DataFrame: _description_
    """

    df_activity_names: pd.DataFrame = pd.read_xml(path_or_buffer = path_file_activity_names)
    df_activity_names: pd.DataFrame = df_activity_names.rename(columns = dict_column_names_activity_names)[dict_column_names_activity_names.values()].dropna().drop_duplicates()

    df_activity_index: pd.DataFrame = pd.read_xml(path_or_buffer = path_file_activity_index)
    df_activity_index: pd.DataFrame = df_activity_index.rename(columns = dict_column_names_activity_index)[dict_column_names_activity_index.values()].dropna().drop_duplicates()

    activities: pd.DataFrame = df_activity_index.merge(
        right = df_activity_names,
        how = 'left',
        on = 'activityNameId'
    )

    return activities
```
